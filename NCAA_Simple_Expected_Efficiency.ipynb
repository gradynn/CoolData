```python
import pandas as pd
```

```python
df = pd.read_csv("stats.csv")
df = new_df = df.drop(index=[363, 364])

for index, row in df.iterrows():
  df.at[index, 'School'] = row['School'].replace('NCAA', '').strip()

df
```

```python
league_eff = df['Eff'].mean()
league_pace = df['Pace'].mean()
print(league_eff)
print(league_pace)
```

```python
df['Off. Diff.'] = df['Eff'] - league_eff
df['Def. Diff.'] = df['Opp. Eff'] - league_eff
df['Pace Diff.'] = df['Pace'] - league_pace
df.head()
```

```python
#
# ENTER TEAMS TO PREDICT BELOW
#
s1 = 'Duke'
s2 = 'Tennessee'

# Find index of school
s1_i = df.loc[df['School'] == s1].index[0]
s2_i = df.loc[df['School'] == s2].index[0]

# Caculate expectations
# eff = (diff to average + opponent diff to average) + average
s1_exp_eff = (df.at[s1_i, 'Off. Diff.'] + df.at[s2_i, 'Def. Diff.']) + league_eff
s2_exp_eff = (df.at[s2_i, 'Off. Diff.'] + df.at[s1_i, 'Def. Diff.']) + league_eff
exp_pace = (df.at[s1_i, 'Pace Diff.'] + df.at[s2_i, 'Pace Diff.']) + league_pace

# Calculate Expected Score
s1_exp_points = (s1_exp_eff / 100) * exp_pace
s2_exp_points = (s2_exp_eff / 100) * exp_pace

# Output
print("--- Expected Values ---")
print(s1 + " expected efficiency: {}".format(s1_exp_eff))
print(s2 + " expected efficiency: {}".format(s2_exp_eff))
print("Expected pace: {}\n".format(exp_pace))
print("--- Expected Result ---")
print(s1 + ": {}".format(s1_exp_points))
print(s2 + ": {}".format(s2_exp_points))
print("- Spread: {}".format(abs(s2_exp_points-s1_exp_points)))
print("- Total: {}".format(s2_exp_points + s1_exp_points))

```
